# PS 8 : Mean-variance investing with fixed, linear-proportional, and quadratic transaction costs and Optimal Dynamic trading of a single asset with quadratic transaction costs


## Problem 1
# 1
To solve the following maximization problem: 
\begin{align}
      \max_{X_1} \left( X_1 \mu-\frac{\gamma}{2} \sigma^2 X_1- \left( \mathbf{1}_{\{X_1 \neq X_0\}} b_0+|X_1-X_0| b_1-\frac{\lambda}{2} (X_1-X_0)^2\right) \right)
\end{align}

we find the first order condition with respect to $X_1$ and we set it equal to zero. \\\\Due to the absolute value in the parenthesis, we have two possible cases:
\begin{align}
    & *when* (X_1-X_0)>0 \text{ the FOC wrt $X_1$ is:}\\
    &\mu-\gamma\sigma^2X_1-b_1-\lambda(X_1-X_0)=0 \longrightarrow X_1 = \frac{\mu+\lambda X_0-b_1}{\gamma\sigma^2+\lambda}\\
    &*when* (X_1-X_0)<0 \text{ the FOC wrt $X_1$ is:}\\
    &\mu-\gamma\sigma^2X_1+b_1-\lambda(X_1-X_0)=0 \longrightarrow X_1 = \frac{\mu+\lambda X_0+b_1}{\gamma\sigma^2+\lambda}
\end{align}
The presence of fixed trading costs and linear proportional trading costs imply that there is a no trade region such that, if $X_0$ is in this region, than for investors is not optimal trading but, instead, they are better off to maintain their dollar position $X_0$.

Due to fixed costs the no trade region is: 
\begin{align}
    \frac{\mu}{\gamma\sigma^2}-\sqrt{\frac{2b_0}{\gamma\sigma^2}}\le X_0 \le \frac{\mu}{\gamma\sigma^2}+\sqrt{\frac{2b_0}{\gamma\sigma^2}}
\end{align}
Due to linear proportional costs the no trade region is: 
\begin{align}
    \frac{\mu-b_1}{\gamma\sigma^2}\le X_0 \le\frac{\mu+b_1}{\gamma\sigma^2}
\end{align}
Thus, the total no trade region is the union of these two regions.
\\In both cases we can rewrite $X_1$ as a convex combination of the $aim$ portfolio and $X_0$: 
\begin{align}
    &*when* (X_1-X_0)>0: X_1 = \frac{\mu-b_1}{\gamma\sigma^2+\lambda}+\frac{\lambda}{\gamma\sigma^2+\lambda}X_0\\
    &\text{then we know that: } 1-\tau = \frac{\lambda}{\gamma\sigma^2+\lambda} \longrightarrow \tau = 1-\frac{\lambda}{\gamma\sigma^2+\lambda}=\frac{\gamma\sigma^2}{\gamma\sigma^2+\lambda}\\
    & \text{and: } \tau\cdot aim =\frac{\mu-b_1}{\gamma\sigma^2+\lambda} \longrightarrow aim = \frac{\mu-b_1}{\gamma\sigma^2} \\\\
    &*when* (X_1-X_0)<0 :\text{ $\tau$ remains the same, while the $aim$ portfolio becomes }\tau \cdot aim =\frac{\mu+b_1}{\gamma\sigma^2+\lambda} \longrightarrow aim = \frac{\mu+b_1}{\gamma\sigma^2}
\end{align}
Summing up: 
\begin{align}
    X_1 = \tau\cdot aim +(1-\tau)X_0\\
\end{align}
where $\tau=\frac{\gamma\sigma^2}{\gamma\sigma^2+\lambda}$ and 
$aim = 
\begin{cases}
\frac{\mu-b_1}{\gamma\sigma^2}, & \text{if } (X_1-X_0)>0 \\
\frac{\mu+b_1}{\gamma\sigma^2}, & \text{if } (X_1-X_0)<0 
\end{cases}$

# 2


*  if $b_0=0$ the trading strategy will not change since $b_0$ is not included either in the $aim$ portfolio nor in $\tau$. The no trade region, instead, changes cause $b_0$ is part of it and, in particular, it becomes narrower.
*  if $b_1=0$ the trading strategy changes both when $(X_1-X_0)>0$ and $(X_1-X_0)<0$, since $b_1$ is part of the $aim$ portfolios; the result is a portfolio that invests $\tau$ in $\frac{\mu}{\gamma\sigma^2}$ and (1-$\tau$) in $X_0$ both if $(X_1-X_0)>0$ and $(X_1-X_0)<0$. The no trade region changes since $b_1$ is part of it and becomes narrower.
*  if $b_0=b_1=0$ the trading strategy change as it does when only $b_1=0$ (point before), while the no trade region becomes narrower with respect of the regions listed before.
*  if $\lambda=0$ and assuming that before was positive, the trading strategy changes because $\tau$ increases. Consequently an investor, both when $(X_1-X_0)>0$ and $(X_1-X_0)<0$, will invest more in the respectively $aim$ portfolio than in $X_0$ than before. The no trade region doesn't change since $\lambda$ doesn't affect it.


## Problem 2

# 1
Given that the value function is given by $V(k,n_{k-1})=\max\limits_{n_t,t\geq k} E \left[ \sum\limits_{t=k}^{T} \rho^{t-k} \left\{ n_t\mu - \frac{\lambda}{2}(n_t-n_{t-1})^2 -\frac{\gamma}{2}n_t^2 \sigma^2 \right\}\right]$, we can write $V(T,n_{T-1})$ as such:
\begin{align}
    V(T,n_{T-1}) &= \max\limits_{n_T,T} E \left[ \sum\limits_{T=k}^{T} \rho^{t-k} \left\{ n_t\mu -\frac{\lambda}{2}(n_t-n_{t-1})^2 -\frac{\gamma}{2}n_t^2 \sigma^2 \right\}\right] = \max\limits_{n_T} \left\{ n_T\mu -\frac{\lambda}{2}(n_T-n_{T-1})^2 -\frac{\gamma}{2}n_T^2 \sigma^2\right\}\\
\end{align}
The, taking the FOC wrt $n_T$:
\begin{align}
    \frac{\partial}{\partial n_T} : \mu - \frac{\lambda }{2} 2(n_T-n_{T-1}) -\frac{\gamma}{2} 2 n_T \sigma^2 = \mu - (\lambda -\gamma \sigma^2)n_T+\lambda n_{T-1} \rightarrow n^*(T,n_{T-1}) = \frac{\mu + n_{T-1}\lambda}{\gamma\sigma^2+\lambda}
\end{align}
Then, plugging $n^*_T$ back into $V(T,n_{T-1})=V(T,n)$ we have:
\begin{align}
    V(T,n) &= n_T\mu -\frac{\lambda}{2}(n_T-n)^2 -\frac{\gamma}{2}n_T^2 \sigma^2\\
    &= \frac{\mu+n\lambda}{\gamma\sigma^2+\lambda}\mu -\frac{\lambda}{2}\left(\frac{\mu+n\lambda}{\gamma\sigma^2+\lambda}-n\right)^2 -\frac{\gamma}{2}\left(\frac{\mu+n\lambda}{\gamma\sigma^2+\lambda}\right)^2 \sigma^2\\
    &=\frac{\mu^2+\mu\lambda n}{\gamma\sigma^2+\lambda}-\frac{\lambda}{2}\left(\frac{\mu+\lambda n-\gamma\sigma^2n-\lambda n}{\gamma\sigma^2+\lambda}\right)^2-\frac{\gamma\sigma^2}{2(\gamma\sigma^2+\lambda)}(\mu+\lambda n)^2\\
    &=\frac{1}{\gamma\sigma^2+\lambda}\left(\mu^2+\mu\lambda n-\frac{\lambda(\mu^2+\gamma^2\sigma^4 n^2-2\mu\gamma\sigma^2n)}{2(\gamma\sigma^2+\lambda)}-\frac{\gamma\sigma^2(\mu^2+\lambda^2 n^2+2\mu\lambda n)}{2(\gamma\sigma^2+\lambda)}\right)
\end{align}
Now we put together the terms that have $n$, $n^2$ and the ones that remain: 
\begin{align}
    &\text{for $n$}: \\\\
    &\left(\mu \lambda+\frac{\lambda \mu \gamma \sigma^2}{\gamma\sigma^2+\lambda}-\frac{\lambda \mu \gamma \sigma^2}{\gamma \sigma^2+\lambda}\right)\frac{n}{\gamma\sigma^2+\lambda}=\frac{\mu \lambda n}{\gamma\sigma^2+\lambda}=q_Tn\\
    &\text{for $n^2$}:\\
    &-\left(\frac{\lambda\gamma^2\sigma^4}{2(\gamma\sigma^2+\lambda)}+\frac{\gamma^2\gamma\sigma^2}{2(\gamma\sigma^2+\lambda)}\right)\frac{n^2}{\gamma\sigma^2+\lambda}
    =-\left(\frac{\lambda\gamma\sigma^2(\gamma\sigma^2+\lambda)}{2(\gamma\sigma^2+\lambda)}\right)\frac{n^2}{\gamma\sigma^2+\lambda}\\
    &=-\frac{\lambda\gamma\sigma^2n^2}{2(\gamma\sigma^2+\lambda)}=-\frac{1}{2}\left(\lambda-\frac{\lambda^2}{\gamma\sigma^2+\lambda}\right)n^2=-\frac{1}{2}Q_Tn^2\\
    &\text{what remains is}:\\
    &\left(\mu^2-\frac{\lambda\mu^2}{2(\gamma\sigma^2+\lambda)}-\frac{\gamma\sigma^2\mu^2}{2(\gamma\sigma^2+\lambda)}\right)\frac{1}{\gamma\sigma^2+\lambda}=\left(\mu^2-\frac{\mu^2(\gamma\sigma^2+\lambda)}{2(\gamma\sigma^2+\lambda)}\right)\frac{1}{\gamma\sigma^2+\lambda}=\\
    &=\frac{\mu^2}{2(\gamma\sigma^2+\lambda)}=c_T
\end{align}
In the end, we arrive to:
\begin{align}
    V(T,n) = -\frac{1}{2}Q_Tn^2+q_Tn+c_T
\end{align}
# 2
Assume that for each $t<T$ the value function is determined by: $V(t+1,n)=-\frac{1}{2}n^2Q_{t+1}+n q_{t+1}+c_{t+1}$.\\
The value function satisfies the Bellman equation: \\\\
$$V(t,n_{t-1})=\max \limits_{n_t} \left(n_t\mu-\frac{\lambda}{2}(n_t-n_{t-1})^2-\frac{\gamma}{2}n_t^2\sigma^2+\rho\mathbb{E}_t[V(t+1,n_t)]\right)$$
Plugging $V(t+1,n)$ in the Bellman equation, we find: 
\begin{align}
    V(t,n_{t-1})=\max \limits_{n_t} \left(n_t\mu-\frac{\lambda}{2}(n_t-n_{t-1})^2-\frac{\gamma}{2}n_t^2\sigma^2+\rho(-\frac{1}{2}n_t^2Q_{t+1}+n_t q_{t+1}+c_{t+1})\right)
\end{align}
taking the FOC wrt $n_t$ and equating it to zero, we find the optimal value of $n_t$, denoted $n^*_t$: 
\begin{align}
    \mu-\lambda(n_t-n_{t-1})-\gamma\sigma^2n_t-\rho Q_{t+1}n_t+\rho q_{t+1}=0 \longrightarrow n_t^* = \frac{\mu+\lambda n_{t-1}+\rho q_{t+1}}{\gamma\sigma^2+\lambda+\rho Q_{t+1}}
\end{align}

Similarly as in question 1, we can plug this back into the value function to write:
\begin{align}
    V(t,n_{t-1})&=n^*_t\mu-\frac{\lambda}{2}(n^*_t-n_{t-1})^2-\frac{\gamma}{2}n_t^{*2}\sigma^2+\rho(-\frac{1}{2}n_t^{*2}Q_{t+1}+n_t^* q_{t+1}+c_{t+1})\\
    &=\frac{n^*_t\mu}{2}+\frac{n^*_t\mu}{2}-\frac{\lambda n^{*2}_t}{2}+\frac{\lambda n^*_t n}{2}+\frac{\lambda n^*_t n}{2} -\frac{n^2\lambda}{2}-\frac{\gamma n^{*2}_t\sigma^2}{2}-\frac{\rho n^{*2}_t Q_{t+1}}{2}+ \frac{n^*_t\rho q_{t+1}}{2}+\frac{n^*_t\rho q_{t+1}}{2} +  \frac{\rho c_{t+1}}{2} +  \frac{\rho c_{t+1}}{2}\\
    &= \frac{1}{2} \left( n^*_t\mu + n^*_t\mu-\lambda n^{*2}_t+\lambda n^*_t n+\lambda n^*_t n -n^2\lambda-\gamma n^{*2}_t\sigma^2-\rho n^{*2}_t Q_{t+1}+ n^*_t\rho q_{t+1}+n^*_t\rho q_{t+1} + \rho c_{t+1} + \rho c_{t+1} \right)\\
    &=\frac{1}{2} \left( n^*_t(\mu -\lambda(n^*_t - n) -\gamma\sigma^2 n^*_t -\rho Q_{t+1} n^*_t ) + n^*_t \mu +\lambda n^*_t n - \lambda n^2 +n^*_t \rho q_{t+1} + \rho c_{t+1} +\rho c_{t+1}\right)\\
    &=\frac{1}{2} (n_t^*\cdot 0 + n^*_t \mu +\lambda n^*_t n - \lambda n^2 +n^*_t \rho q_{t+1} + \rho c_{t+1} +\rho c_{t+1})\\
    &=-\frac{\lambda}{2}n^2+\frac{n^*_t \lambda }{2}n + \frac{(\mu+\rho q_{t+1})n^*_t}{2} + \rho c_{t+1}\\
    &=-\frac{\lambda}{2}n^2 +\frac{\frac{\mu+\lambda n+\rho q_{t+1}}{\gamma\sigma^2+\lambda+\rho Q_{t+1}} \lambda }{2}n + \frac{(\mu+\rho q_{t+1})\frac{\mu+\lambda n+\rho q_{t+1}}{\gamma\sigma^2+\lambda+\rho Q_{t+1}}}{2} + \rho c_{t+1}\\
    &= -\frac{1}{2}\lambda(1-\frac{\lambda}{\gamma\sigma^2+\lambda+\rho Q_{t+1}})n^2+ \frac{\lambda(\mu+\rho q_{t+1})}{\gamma\sigma^2+\lambda+\rho Q_{t+1}} n + \frac{(\mu+\rho q_{t+1})^2}{2(\gamma\sigma^2+\lambda+\rho Q_{t+1})} + \rho c_{t+1}
\end{align}

Thus if we define 
\begin{align}
    Q_t &= \lambda(1-\frac{\lambda}{\gamma\sigma^2+\lambda+\rho Q_{t+1}})\\
    q_t &= \frac{\lambda(\mu+\rho q_{t+1})}{\gamma\sigma^2+\lambda+\rho Q_{t+1}}\\
    \text{and } c_t &= \frac{(\mu+\rho q_{t+1})^2}{2(\gamma\sigma^2+\lambda+\rho Q_{t+1})} + \rho c_{t+1},
\end{align}
using these recursively defined parameters, we can write $V(t,n) = -\frac{1}{2}n^2 Q_t + q_t n + c_t$.
# 3
Using the fact that $aim_t$ maximizes the value function $V(t,n) = -\frac{1}{2}n^2 Q_t + q_t n + c_t$ at any time $t$, we know that the first derivative wrt $n$ of $V(t,n)$ must equal 0:
\begin{align}
    \frac{\partial V(t,n)}{\partial n} \Bigr|_{aim_t}= 0 = -Q_t aim_t + q_t \rightarrow aim_t = \frac{q_t}{Q_t}
\end{align}
Plugging in $Q_t$ and $q_t$ as found in question 2, we have:
\begin{align}
    aim_t=\frac{\mu +\rho q_{t+1}}{\gamma \sigma^2 +\rho Q_{t+1}}
\end{align}
Then, rewriting $n_t^* = \frac{\mu+\lambda n_{t-1}+\rho q_{t+1}}{\gamma\sigma^2+\lambda+\rho Q_{t+1}}$:
\begin{align}
    n_t^* &= \frac{\mu+\rho q_{t+1}}{\gamma\sigma^2+\rho Q_{t+1}+\lambda}+\frac{\lambda n_{t-1}}{\gamma\sigma^2+\lambda+\rho Q_{t+1}}\\
    &= \frac{\mu+\rho q_{t+1}}{\gamma\sigma^2+\rho Q_{t+1}+\lambda}\cdot \frac{\gamma\sigma^2+\rho Q_{t+1}}{\gamma\sigma^2+\rho Q_{t+1}} + \frac{\lambda}{\gamma\sigma^2+\lambda+\rho Q_{t+1}}n_{t-1}\\
    &= aim_t \cdot \frac{\gamma\sigma^2+\rho Q_{t+1}}{\gamma\sigma^2+\rho Q_{t+1}+\lambda} + n_{t-1}(\frac{\gamma\sigma^2+\rho Q_{t+1}+\lambda}{\gamma\sigma^2+\rho Q_{t+1}+\lambda} - \frac{\gamma\sigma^2+\rho Q_{t+1}}{\gamma\sigma^2+\rho Q_{t+1}+\lambda}) 
\end{align}
If we define $\frac{\gamma\sigma^2+\rho Q_{t+1}}{\gamma\sigma^2+\rho Q_{t+1}+\lambda} = \tau_t$, we can write
\begin{align}
    n^*_t =  \tau_t \cdot aim_t + (1-\tau_t) \cdot n_{t-1}
\end{align}
as required.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Constants
mu      = 0
gamma   = 1
n_start  = 1000000
sigma_a = 0.3 
day_start = '09:30:00'
day_end   = '16:00:00'
```

```python
def timestamp_trade(time_start,time_end,freq_min): 
    trade_time  = pd.date_range(time_start,time_end,freq="%smin"%freq_min)
    return trade_time
```

```python
def transaction_costs(timestamp,lunch_start,lunch_end,pic,pic_lunch): 
    pic/=10000
    pic_lunch/=10000
    pic_out=pd.DataFrame({'pic': pic}, index = timestamp)
    pic_out[(pic_out.index.hour >= lunch_start) & (pic_out.index.hour <= lunch_end)] = pic_lunch
    pic_out=pic_out.to_numpy()
    return pic_out
```

```python
def trading_volume(mu,gamma,n_start,sigma_a,trade_time,lunch_start,lunch_end,pic,pic_lunch):
    
    #Trade schedule volatility, & t-costs
    N              = len(trade_time)
    sigma          = sigma_a/np.sqrt(252*(N-1))
    pic            = transaction_costs(trade_time,lunch_start,lunch_end,pic,pic_lunch)

    #Coefficients at time t=T+1,T-1,...,1
    Q    = np.ones((N+1,1))
    q    = np.ones((N+1,1))
    C    = np.ones((N+1,1))
    Q[N] = 0
    q[N] = 0
    C[N] = 0
    for t in range(N-1,0,-1): 
        Q[t] = pic[t]*(1-pic[t]/(Q[t+1]+gamma*sigma**2+pic[t]))
        q[t] = pic[t]*(q[t+1]+mu)/(Q[t+1]+gamma*sigma**2+pic[t])
        C[t] = C[t+1]+0.5*(q[t+1]+mu)**2/(Q[t+1]+gamma*sigma**2+pic[t])

    #Optimal trading at time t=0,1,2,...T    
    tau      = np.divide(Q[:-1],pic)
    aim      = np.divide(q[:-1],Q[:-1])
    n_share      = np.empty((N,1))
    t_cost    = np.empty((N,1))
    n_share[0]   = n_start
    t_cost[0] = 0
    for t in range(1,N,1): 
        n_share[t]   = tau[t]*aim[t]+(1-tau[t])*n_share[t-1]
        t_cost[t] = t_cost[t-1]+0.5*pic[t]*(n_share[t]-n_share[t-1])**2
    return n_share,t_cost
```

```python
#30 min 
t30 = timestamp_trade(day_start,day_end,30)
n30,tc30=trading_volume(mu,gamma,n_start,sigma_a,t30,12,14,1,1)
#10 min 
t10 = timestamp_trade(day_start,day_end,10)
n10,tc10=trading_volume(mu,gamma,n_start,sigma_a,t10,12,14,1,1)


plt.plot(t30,n30,'r*-',label="30 minutes")
plt.plot(t10,n10,'b.-',label="10 minutes")
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%H:%M"))
plt.legend(loc='upper right')
plt.xlabel('Time (hour)')
plt.ylabel('# of shares held')
plt.show() 
```

```python
plt.plot(t30,tc30,'r*-',label="30 minutes TC")
plt.plot(t10,tc10,'b.-',label="10 minutes TC")
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%H:%M"))
plt.legend(loc='upper right')
plt.xlabel('Time (hour)')
plt.ylabel('cumulative trading costs')
plt.show() 
```

# 5

The trade-off between mean return and risk at a given time, t, is solely determined by the position at that same moment, due to the constancy of both the mean and variance. The historical price trends do not affect the investor's balancing act between the stock's potential return and its associated risk, nor does it influence the liquidation cost of the stock.

This may seem contradictory at first glance as one might naturally assume that the sale price, even when taking trading costs into account, significantly impacts the day's outcome. However, within this model, only the transaction cost influences the investor's cost function. As these costs are symmetric for both buying and selling, the optimal trading plan does not hinge on realized price shocks.

This model could be adjusted to better reflect reality in at least two ways. Firstly, we could consider that investors' risk aversion fluctuates with their wealth. If they are more cautious when their wealth is relatively lower, they may initially own more shares than desired and would be willing to sell more towards their target position following an initial decline in the share price.

Secondly, if transaction costs are dependent on the state, the model would not be applicable. For example, if a falling market leads to a liquidity squeeze, the investor may face higher transaction costs if the stock price drops. Alternatively, if transaction costs are tied to the trading volume rather than the quantity of shares traded, transaction costs could escalate if the stock price rises.

$\lambda$ affects $Q_t,q_t,c_t$ and $n_t$ for each $t$; consequently, if $\lambda$ changes depending on the time, to compute the above parameters, we need to control at which time we are. If time $t$ is between 12:00pm and 02:00pm, then we will set $\lambda_t=2bps$, otherwise we will use $\lambda_t=1bps$. In this case, once we are in the lunch break, the investor will trade less than in the morning in order to not pay an high volume of transaction costs. We can clearnly see the difference in shares held in the plot of differences (order of magnitude of a couple of 1000), but in the number of shares held, it's not obvious that there is a large change because we are looking at an order of magnitude in the few millions of shares.
