## Point (a)
Each investor solve the following problem:
$$\begin{aligned}\max_{w^i}(\mathbb{E}[R_p^i]-\frac{a^i}{2}V[R_p^i]) \text{  such that  }(w^i)^T\textbf{1}=1\\\Longrightarrow 
\max_{w^i}((w^i)^T\mu-\frac{a^i}{2}(w^i)^T\Sigma(w^i)) \text{  such that  }(w^i)^T\textbf{1}=1\end{aligned}$$
Setting the langrangian and setting the first order condition wrt $w^i$ equal to zero, we have: 
$$\begin{aligned}L=((w^i)^T\mu-\frac{a^i}{2}(w^i)^T\Sigma(w^i))+\lambda^i(1-(w^i)^T\textbf{1})\Longrightarrow \frac{dL}{dw^i}=\mu-a^i\Sigma w^i-\lambda^i\textbf{1}=0\\ 
w^i=\frac{1}{a^i}\Sigma^{-1}(\mu-\lambda^i\textbf{1})\text{ with }\lambda^i=\frac{\textbf{1}\Sigma^{-1}\mu-a^i}{\textbf{1}\Sigma^{-1}\textbf{1}}\end{aligned}$$
We found $\lambda^i$ by substituting back the optimal $w^i$ in the constraint.
Since $w_{slope}=\frac{\Sigma^{-1}\mu}{B}$ and $w_{min}=\frac{\Sigma^{-1}\textbf{1}}{A}$, we can write $w^i=\frac{B}{a^i}w_{slope}+(1-\frac{B}{a^i})w_{min}$ with $B=\textbf{1}^T\Sigma^{-1}\mu$ and $A=\textbf{1}^T\Sigma^{-1}\textbf{1}$ .

Then, 
$$\begin{aligned}\sum_{i=1}^{I}w^i\frac{W^i}{W}=\sum_{i=1}^{I}\frac{1}{a^i}\frac{W^i}{W}\Sigma^{-1}(\mu-\lambda^i\textbf{1})\Longrightarrow w^m=\frac{1}{a}\Sigma^{-1}(\mu-\lambda^i\textbf{1})=\frac{B}{a}w_{slope}+(1-\frac{B}{a})w_{min}\end{aligned}$$
The market portfolio is mean variance efficient becuase it lies on the efficient frontier since is the solution of the following maximization problem: 
$$\begin{aligned}\max_{w^i}(\mathbb{E}[R_p^i]-\frac{a}{2}V[R_p^i]) \text{  such that  }(w^i)^T\textbf{1}=1\\\Longrightarrow 
\max_{w^i}((w^i)^T\mu-\frac{a}{2}(w^i)^T\Sigma(w^i)) \text{  such that  }(w^i)^T\textbf{1}=1\end{aligned}$$
To derive the security market line we need to find a ralation between the expected return and the beta of each asset. Let $A=\textbf{1}^T\Sigma^{-1}\textbf{1}$.

$$\begin{aligned}Cov(R,R_m)=\Sigma w_m=\Sigma\left(\frac{B}{a}w_{slope}+\left(1-\frac{B}{a}\right)w_{min}\right)=\frac{\mu}{a}+\frac{1}{A}\left(1-\frac{B}{a}\right)\textbf{1}\\
\Longrightarrow \mu = a\left(Cov(R,R_m)-\frac{1}{A}\left(1-\frac{B}{a}\right)\textbf{1}\right)\end{aligned}$$
For the market portfolio we have: $$\sigma_m^2=w_m^T\Sigma w_m=\frac{1}{a}\mu_m+\left(1-\frac{1}{a}\right)\frac{1}{A}$$
We take another minimum variance portfolio $w_z$ that is uncorrelated with $w_m$: $$w_z^T\Sigma w_m=\frac{1}{a}\mu_z+\left(1-\frac{1}{a}\right)\frac{1}{A}=0$$
Then we solve these equations to find $\mu$, ending with: $$\mu=\mu_z\textbf{1}+\frac{Cov(R,R_m)}{\sigma_m^2}(\mu_m-\mu_z)=\mu_z\textbf{1}+\beta(\mu_m-\mu_z)$$
If $\mu_z>R_f$ then the security market like in this setting is flatter than the security market like in the usual CAPM.

If for each investor, borrowing at the risk free rate is constrained to a fraction $l^i$, then we can write $(w^i)^T\textbf{1}<=1+l^i$, that is: the amount invested in the risky assets can't go above your initial wealth (1) plus an additional fraction of wealth ($l^i$) that you can borrow at the risk free rate.\
Now investors' optimization problem is of the following form: 
$$\begin{aligned}\max_{w^i}(\mathbb{E}[R_p^i]-\frac{a^i}{2}V[R_p^i])=\max_{w^i}(R_f+(w^i)^T(\mu-R_f\textbf{1})-\frac{a^i}{2}(w^i)^T\Sigma w^i) \text{  such that  }\frac{(w^i)^T\textbf{1}}{1+l^i}<=1\\
\Longrightarrow L=R_f+(w^i)^T(\mu-R_f\textbf{1})-\frac{a^i}{2}(w^i)^T\Sigma w^i+\lambda^i\left(\frac{(w^i)^T\textbf{1}}{1+l^i}\right)\Longrightarrow 
\frac{dL}{dw^i}=\mu-R_f\textbf{1}-a^i\Sigma w^i-\frac{\lambda^i\textbf{1}}{1+l^i}=0\end{aligned}$$
Then we have: 
$$\begin{aligned} w^i=\frac{1}{a^i}\Sigma^{-1}\left(\mu-R_f\textbf{1}-\frac{\lambda^i\textbf{1}}{1+l^i}\right)
\end{aligned}$$
The constrained investor invests in portfolios that are convex combination of $w_{tan},w_{min}$ and $R_f$; this is the three funds separation theorem.



## Point (c)
Imposing the market clearing we can write:
$$\begin{aligned} w_m=\sum_{i=1}^I\left(\frac{W^i}{W}w^i\right)=
\sum_{i=1}^I\left(\frac{W^i}{W}\frac{1}{a^i}\right)\Sigma^{-1}\left(\mu-R_f\textbf{1}-\sum_{i=1}^I\left(\frac{W^i}{W}\frac{a}{a^i}\lambda^i\frac{1}{1+l^i}\right)\textbf{1}\right)=\frac{1}{a}\Sigma^{-1}(\mu-R_f\textbf{1}-\psi\textbf{1}) \end{aligned}$$
Now we have: 
$$\begin{aligned} Cov(R_n,R_m)=(\Sigma w_m)_{n,n}=\frac{1}{a}(\mu_n-R_f-\psi) \text{ and } Var(R_m)=w_m^T\Sigma w_m=\frac{1}{a}(\mu_m-R_f-\psi)\\
\Longrightarrow \mu_n=R_f+\psi+\frac{Cov(R_n,R_m)}{Var(R_m)}(\mu_m-R_f-\psi)=(R_f+\psi)+\beta_n(\mu_m-R_f-\psi)\end{aligned}$$

## Point (c): Constructing the BAB factor

```python
df1 = data.copy()
df1.sort_values(['date','beta'], inplace=True)
df1.dropna(inplace=True)
df1
```

```python
df1['z'] = df1.groupby('date')['beta'].rank()
df1['z_mean'] = df1.groupby('date')['z'].transform(lambda x: np.mean(x))
df1['diff'] = df1['z']-df1['z_mean']
df1['k'] = df1.groupby('date')['diff'].transform(lambda x: 2/np.sum(np.abs(x)))
df1['wh'] = df1['diff'] * df1['k']
df1['wh'] = df1['wh'].clip(lower=0)
df1['wl'] = df1['diff'] * df1['k']
df1['wl'] = df1['wl'].clip(upper=0)
df1['wl'] = df1['wl'].abs()
df1['z']
```

```python
df1['Rh'] = df1['Rn']*df1['wh'].shift(1)
df1['Rh'] = df1.groupby('date')['Rh'].transform(lambda x: np.sum(x))
df1['Rl'] = df1['Rn']*df1['wl'].shift(1)
df1['Rl'] = df1.groupby('date')['Rl'].transform(lambda x: np.sum(x))
df1['beta_h'] = df1['wh']*df1['beta']
df1['beta_h'] = df1.groupby('date')['beta_h'].transform(lambda x: np.sum(x))
df1['beta_l'] = df1['wl']*df1['beta']
df1['beta_l'] = df1.groupby('date')['beta_l'].transform(lambda x: np.sum(x))
df2 = df1.copy()
df2.drop_duplicates(subset=['date'],keep='first', inplace=True)
```

```python
df2['BL_l'] = df2['beta_l'].shift(1)
df2['BH_l'] = df2['beta_h'].shift(1)
df2['R_bab'] = (df2['Rl']-df2['rf'])/df2['BL_l']-(df2['Rh']-df2['rf'])/df2['BH_l']
df2['Re_bab'] = df2['R_bab'] - df2['rf']

# sharpe ratio of the BAB factor
sharpe_ratio = (df2['R_bab'].mean()*12)/(df2['R_bab'].std()*np.sqrt(12))
print('The sharpe ratio of the BAB factor is:', round(sharpe_ratio,3))
```

```python
df2.dropna(inplace=True)
df2.drop(columns=['z','z_mean','diff','k','wh','wl','Rh','Rl','beta_h','beta_l','BL_l','BH_l','prc','shrout','Rn','permno', 'Unnamed: 0' , 'mcap', 'mcap_l', 'Rn_e', 'w_m','N','const'], inplace=True)
df2
```

```python
# CAPM alpha 
y = df2['R_bab']
x = df2['Rm_e']

x = sm.add_constant(x)
regression = sm.OLS(y, x).fit()
alpha = regression.params[0]
beta = regression.params[1]
print('The CAPM alpha of the BAB factor is:', round(alpha,3))
print('The CAPM beta of the BAB factor is:', round(beta,3))
```

## Point (d): Long-Short Portfolios

Long-Short Portfolios RBAB is the return to a self-financing (i.e. dollar neutral) t+1
long short portfolio. Is RBAB also dollar neutral in terms of risky securities only? t+1
Intuitively, describe the difference to the portfolio RH − RL.\
A self-financing long-short portfolio is designed to be dollar neutral, meaning that the total amount of money invested in long positions is equal to the total amount of money invested in short positions. This ensures that the portfolio's overall exposure to the market is neutral and not dependent on the direction of the market.

However, this dollar neutrality applies to the overall portfolio, including both risky and risk-free securities. It does not necessarily mean that the portfolio is dollar neutral in terms of risky securities only.

The difference between the portfolio RH − RL and R_BAB is that RH − RL represents the return difference between a long portfolio and a short portfolio, while R_BAB is the return of a self-financing long-short portfolio. RH and RL are not necessarily dollar-neutral portfolios, whereas R_BAB is designed to be dollar neutral.

In other words, RH and RL may have different levels of exposure to the market, while R_BAB is constructed to have zero net exposure. Therefore, R_BAB is a more controlled and consistent investment strategy that is not affected by the overall market direction, whereas RH − RL is more dependent on market trends and the specific long and short positions taken by the investor.

## Point (e): Long-Short Portfolios

```python
beta_BAB=beta
alpha_BAB=alpha

# Define the objective function to be minimized
def objective_function(x, df):
    alloc_market, alloc_BAB, alloc_rf = x[0], x[1], x[2]
    portfolio_return = alloc_market * df['Rm'] + alloc_BAB * df['R_bab'] + alloc_rf * df['rf']
    portfolio_volatility = np.sqrt(alloc_market**2 * df['Rm'].var() + alloc_BAB**2 * df['R_bab'].var() + alloc_rf**2 * df['rf'].var())
    sharpe_ratio = -1 * (portfolio_return- df['rf']).mean() / portfolio_volatility
    return sharpe_ratio

# Define the equality constraint
def equality_constraint(x):
    return np.sum(x) - 1

# Define the bounds for the optimization
bounds = ((-3, 3), (-3, 3), (-3, 3))

# Solve for the optimal allocation
result = minimize(objective_function, [0.5, 0.5, 0.0], method='SLSQP', constraints={'type': 'eq', 'fun': equality_constraint}, bounds=bounds, args=(df2,))
optimal_allocations = result.x

# Calculate portfolio statistics for optimal allocation
optimal_return = optimal_allocations[0] * df2['Rm'] + optimal_allocations[1] * df2['R_bab'] + optimal_allocations[2] * df2['rf']
optimal_volatility = np.sqrt(optimal_allocations[0]**2 * df2['Rm'].var() + optimal_allocations[1]**2 *  df2['R_bab'].var() + optimal_allocations[2]**2 * df2['rf'].var())
optimal_sharpe_ratio = -1 * objective_function(optimal_allocations, df2)

# Print the optimal allocation and portfolio statistics
print("Optimal allocation to risk-free rate:", optimal_allocations[2])
print("Optimal allocation to market portfolio:", optimal_allocations[0])
print("Optimal allocation to BAB factor:", optimal_allocations[1])
print("Portfolio return:", optimal_return.mean())
print("Portfolio volatility:", optimal_volatility)
print("Portfolio Sharpe ratio:", optimal_sharpe_ratio)

```

When the funding constraint is tight, it means that investors face higher transaction costs and have limited access to funding to invest in high-beta stocks. This causes high-beta stocks to be relatively overpriced, while low-beta stocks are relatively underpriced. The BAB strategy takes advantage of this mispricing by taking long positions in low-beta stocks and short positions in high-beta stocks. As a result, the BAB return tends to be higher when the funding constraint is tight.

Conversely, when the funding constraint is loose, investors have lower transaction costs and easier access to funding to invest in high-beta stocks. This leads to relatively lower prices for low-beta stocks and higher prices for high-beta stocks, reducing the attractiveness of the BAB strategy. Therefore, the BAB return tends to be lower when the funding constraint is loose.

In summary, the relationship between the BAB return and the aggregate tightness of the funding constraint ψ is that a tight funding constraint tends to be associated with higher BAB returns, while a loose funding constraint tends to be associated with lower BAB returns.
