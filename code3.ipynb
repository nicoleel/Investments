# Problem 1

## part a)
An optimal portfolio weight vector $\vec{w}$ is such that for the
corresponding mean-variance efficient portfolio return Rp we have:
$$\begin{aligned}
\vec{w}^*&=argmax\left\{ \mathbb{E}(R_p)-\frac{\gamma}{2}Var{R_p} \right\}\\
       &=argmax\left\{ \sum_{i=0}^nw_i\mathbb{E}(R_i)-\frac{\gamma}{2}Var(\sum_{i=0}^nw_iR_i)\right\}\\
       &=argmax\left\{w_0R_f+\sum_{i=1}^nw_i\mu_i-\frac{\gamma}{2}Var({\sum_{i=1}^nw_iR_i)}\right\}\\
       &=argmax\left\{(1-\sum_{i=1}^nw_i)R_f+\sum_{i=1}^nw_i\mu_i-\frac{\gamma}{2}\sum_{j=1}^n\sum_{i=1}^nw_i\Sigma_{ij}w_j \right\}
\end{aligned}$$
with the constraint: $\sum_{i=0}^nw_i=1$, that is
$w_0=1-\sum_{i=1}^nw_i$, the weight associated to the constant risk free
asset $R_f$  
Using the symmetry of the covariance matrix $\Sigma$, the first order
condition is thus: $$\begin{aligned}
    \mu_i-R_f-\gamma\sum_{i=1}^n\Sigma_{ij}w_j^*=0\Leftrightarrow \mu_i-R_f=\gamma cov[R_i,R_p] \, \forall i\in \{1,2,..., n\}\end{aligned}$$

## part b)

From a), we have $$\begin{aligned}
    \vec{\mu}-R_f\vec{1}=\gamma\Sigma w&\Rightarrow w^t(\vec{\mu}-R_f\vec{1})=\gamma w^t\Sigma w\\
    &\Rightarrow \sum_{i=1}^nw_i \mu_i-\sum_{i=1}^n w_iR_f=\gamma \sum_{i=1}^n\sum_{j=1}^nw_i\Sigma_{ij} w_j\\
    &\Rightarrow \mu_p-R_f=\gamma\sigma_p^2\end{aligned}$$ where we used
that $w_0=1-\sum_{i=1}^nw_i$

Thus $$\begin{aligned}
    \beta_{i,p}=\frac{Cov[R_i,R_p]}{\sigma_p^2}&\Leftrightarrow \beta_{i,p}=\frac{\mu_i-R_f}{\gamma\sigma_p^2} \text{      (by point a)}\\
    &\Leftrightarrow \beta_{i,p}=\frac{\mu_i-R_f}{\mu_p-R_f} \text{      (by the previous calculation)}\end{aligned}$$

## Part c)

The model for the linear regression of $R_i$ on $R_p$ is given by:

$$ \begin{equation}
    R_i=\beta_0+\beta_1 R_p +\epsilon_i
    \end{equation}$$
    
where $\epsilon_i$ is the error term with expectation 0. The coefficient $\beta_1$ is given by:
$$\beta_1=\frac{Cov[R_i,R_p]}{\sigma_p^2}=\beta_{i,p}$$ 

Taking the
expectation on both side of $(1)$ gives: $$\begin{aligned}
    \mu_i=\beta_0+\beta_{i,p}\mu_p\Leftrightarrow \beta_0=\mu_i-\beta_{i,p}\mu_p\end{aligned}$$
Injecting this result in $(1)$
 and using that from b) we have
$$\begin{aligned}
\mu_i=R_f+\beta_{i,p}(\mu_p -R_f)
\end{aligned} $$
We have:

$$\begin{aligned}
    R_i&=R_f+\beta_{i,p}(\mu_p -R_f)-\beta_{i,p}\mu_p+\beta_{i,p}R_p+\epsilon_i\\
    &=R_f+\beta_{i,p}(R_p-R_f)+\epsilon_i\end{aligned}$$
 

## part d)

Consider two mean-variance efficient portfolios $R_p$ and $R_q$. We know
that by two fund separation, they can be written as a combination of the
tangency portfolio $R_{tan}$ and the risk free asset $R_f$, that is we
can write: $$\begin{aligned}
    R_p=w_pR_{tan}+(1-w_p)R_f\\
    R_q=w_qR_{tan}+(1-w_q)R_f\end{aligned}$$ Now, from point b) we have:
$$\begin{aligned}
    \frac{\mu_p-R_f}{\sigma_p}&=\sigma_p \frac{\mu_{tan}-R_f}{Cov[R_{tan},R_p]}\\
    &= \sigma_q\frac{\mu_{tan}-R_f}{Cov[R_{tan},R_q]} \Bigl( \frac{Cov[R_{tan},R_q]}{Cov[R_{tan},R_p]}\frac{\sigma_p}{\sigma_q} \Bigr) \\
    &= \frac{\mu_q-R_f}{\sigma_q} \Bigl(\frac{\sigma_pw_qVar(R_{tan})}{\sigma_qw_pVar(R_{tan})}\Bigr)\\
    &= \frac{\mu_q-R_f}{\sigma_q} \Bigl(\frac{w_q \sigma_{tan} w_p}{w_p \sigma_{tan} w_q}\Bigr)\\
     &= \frac{\mu_q-R_f}{\sigma_q}
\end{aligned}$$ 

which shows that
Sharpe Ratio is constant over mean-variance efficient portfolio.



# Problem 2

## (a)
Suppose we have N risky assets.\
$\mu$ is the column vector with N element containing the mean of each risky asset.\
$\Sigma$ is the NxN covariance matrix.\
$\mu_p$ and $\sigma_p^2$ are respectively the mean and the variance of the portfolio composed by N risky assets.

The minimum variance portfolio frontier is the solution **w** to the following minimization problem:
$$\begin{align*} \min\frac{\textbf{w}'\Sigma\textbf{w}}{2}\text{  such that  }\textbf{1}'\textbf{w}=1\text{  and  } \mu'\textbf{w}=\mu_p\end{align*}$$
The lagrangian of the problem is: 
$$\begin{align*} L=\frac{\textbf{w}'\Sigma\textbf{w}}{2}+\lambda(1-\textbf{1}'\textbf{w})+\gamma(\mu_p-\mu'\textbf{w})\end{align*}$$
and the consequent first order conditions are: 
$$\begin{cases}
  \frac{\partial{L}}{\partial\textbf{w}} &= \Sigma\textbf{w}-\lambda\textbf{1}-\gamma\mu=0 \\
  \frac{\partial{L}}{\partial\lambda} &= 1-\textbf{1}'\textbf{w}=0 \\
  \frac{\partial{L}}{\partial\gamma} &= \mu_p-\mu'\textbf{w}=0
\end{cases}$$

All minimum variance portfolios, depending on the two lagrange multipliers $\lambda$ and $\gamma$, are:
$$\begin{align*} \textbf{w} = \lambda\Sigma^{-1}\textbf{1}+\gamma\Sigma^{-1}\mu\end{align*}$$ 
We can notice that $\textbf{w}$ is described by two Nx1 vectors.\
Using the two constraint, we can find $\lambda$ and $\gamma$: 
$$\begin{align*} \lambda= \frac{\mu'\Sigma^{-1}\mu-\mu_p\textbf{1}'\Sigma^{-1}\mu}{\textbf{1}'\Sigma^{-1}\textbf{1}\cdot\mu'\Sigma^{-1}\mu-(\textbf{1}'\Sigma^{-1}\mu)^2}\text{ and } \gamma= \frac{\mu_p\textbf{1}'\Sigma^{-1}\textbf{1}-\textbf{1}'\Sigma^{-1}\mu}{\textbf{1}'\Sigma^{-1}\textbf{1}\cdot\mu'\Sigma^{-1}\mu-(\textbf{1}'\Sigma^{-1}\mu)^2}\end{align*}$$
Letting:\
$A=\textbf{1}'\Sigma^{-1}\textbf{1}$\
$B=\textbf{1}'\Sigma^{-1}\mu$\
$C=\mu'\Sigma^{-1}\mu$\
$\Delta=\textbf{1}'\Sigma^{-1}\textbf{1}\cdot\mu'\Sigma^{-1}\mu-(\textbf{1}'\Sigma^{-1}\mu)^2=AC-B^2$\
and substituting $\lambda$ and $\gamma$ in the expression for **w**, we find: 
$$\begin{align*} \textbf{w} = \frac{1}{\Delta}(C-\mu_pB)\Sigma^{-1}\textbf{1}+\frac{1}{\Delta}(\mu_pA-B)\Sigma^{-1}\mu\end{align*}$$
Now, if we multiply and divide the fist term by A and the second term by B, we have:
$$\begin{align} \textbf{w} &= \frac{1}{\Delta}(C-\mu_pB)A\frac{\Sigma^{-1}\textbf{1}}{A}+\frac{1}{\Delta}(\mu_pA-B)B\frac{\Sigma^{-1}\mu}{B}\\
&=\frac{1}{\Delta}(C-\mu_pB)A\textbf{w}_{min}+\frac{1}{\Delta}(\mu_pA-B)B\textbf{w}_{slope}\end{align}$$
$\textbf{w}_{min}$ and $\textbf{w}_{slope}$ are two particular minimum variance frontier portfolios. The former is the one that has the smallest variance possible, the latter is the portfolio that maximizes the slope of a straight line between the origin and a point on the minimum variance frontier.\
Using the notation of the assignment:\
$\alpha=\frac{1}{\Delta}(C-\mu_pB)A$\
$1-\alpha=\frac{1}{\Delta}(\mu_pA-B)B$\
$\textbf{w}_{min}=\textbf{w}_a$\
$\textbf{w}_{slope}=\textbf{w}_b$



## (b)

Given that the constructed portfolio consists of a fraction $w$ in the risky asset with retrun $R$ and a fraction $(1-w)$ in the global minimum variance portfolio of risky assets with return $R_{min}$, then by construction the variance of the obtained portfolio is minimized at $w=0$. To show this, we have:\
$$
var(P)=var(w R + (1-w) R_{min})=w^2\sigma_R^2+(1-w)^2\sigma_{min}^2+2w(1-w)cov(R,R_{min})\\
\frac{\partial var(P)}{\partial w}=2w\sigma_R^2-2(1-w)\sigma_{min}^2+2cov(R,R_{min})-4w cov(R,R_{min})=2w(\sigma_R^2+\sigma_{min}^2-2cov(R,R_{min}))+2cov(R,R_{min})-2\sigma_{min}^2=0\\
\frac{\partial^2 var(P)}{\partial w^2}=2(\sigma_R^2+\sigma_{min}^2-2cov(R,R_{min}))=2var(R-R_{min})>0
$$
because a variance is always $>0$. Thus, we have that $w=0$ is minimizes the variance of the portfolio, from which we can rewrite the variance as:
$$
var(0* R + (1-0) * R_{min})=var(R_{min})=0^2\sigma_R^2+(1-0)^2\sigma_{min}^2+2*0*(1-0)cov(R,R_{min})=cov(R,R_{min})\\
var(R_{min}) = cov(R,R_{min})
$$

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
from tabulate import tabulate
import wrds
```

```python
db = wrds.Connection(wrds_username='nicoles')
```

```python
stocks = db.raw_sql("select date, vwretd from crsp.msi where date>='1960-01-01' and date<='2023-01-01'", date_cols=['date'])
bonds = db.raw_sql("select caldt, b2ret from crsp.mcti where caldt>='1960-01-01' and caldt<='2023-01-01'", date_cols=['caldt'])
```

```python
stocks.set_index('date', inplace=True)
bonds.set_index('caldt', inplace=True)
data = pd.concat([stocks.rename(columns={"vwretd":"sret"}),bonds.rename(columns={"b2ret":"bret"})],axis=1)
```

## (a)

```python
weight_sb = np.array([[0.6], [0.4]])

#annualized means, variances, covariances
mean_sb = data.mean() * 12
std_sb = data.std() * np.sqrt(12)
var_sb = data.var() * 12
cov_sb = data.sret.cov(data.bret) * 12
corr_sb = data.sret.corr(data.bret) 

print("mean stocks return:", round(mean_sb[0],4), "\nmean bonds return:", round(mean_sb[1],4))
```

```python
mean_6040 = (weight_sb[0]*mean_sb[0] + weight_sb[1]*mean_sb[1]).item()
var_6040 = (weight_sb[0]**2 * var_sb[0] + weight_sb[1]**2 * var_sb[1] + 2*weight_sb[0]*weight_sb[1]*cov_sb).item()
std_6040 = np.sqrt(var_6040)
print("60/40 portfolio mean:", round(mean_6040,4),"\n60/40 portfolio stdev:", round(std_6040,4))
```

```python
rf = 0.01
# annualized Sharpe ratio
sharpe_6040 = (mean_6040 - rf)/std_6040
print("60/40 portfolio Sharpe ratio:", round(sharpe_6040,4))
```

The tangency portfolio weight is given by :
    $w_{tan}=\frac{\Sigma^{-1}\mu}{1^T \Sigma^{-1} \mu}$

```python
weights = np.linspace(start=0, stop=1.01, num=1000)
expected_return = weights * mean_sb[0] + (1-weights) * mean_sb[1]
standard_deviation = np.sqrt(weights ** 2 * var_sb[0] + (1 - weights) ** 2 * var_sb[1] + 2 * weights * (1-weights) * cov_sb)
sharpe_ratio = (expected_return - rf) / standard_deviation 
sharpe_ratio_tan = sharpe_ratio.max()
index_tan = np.argmax(sharpe_ratio)
mean_tan = expected_return[index_tan]
std_tan = standard_deviation[index_tan]
weight_tan = weights[index_tan]
print("The tangency portfolio has weights\n",
      "w_s (stocks):",round(weight_tan,4), "\n",
      "w_b (bonds) :",round(1-weight_tan,4))

print("The tangency portfolio has:\n", 
      "mean        :", round(mean_tan,4),"\n",
      "std dev     :", round(std_tan,4),"\n", 
      "Sharpe ratio:", round(sharpe_ratio_tan,4))
```

```python
weights_short = np.linspace(start=-0.7, stop=1.5, num=1000)
expected_return_short = weights_short * mean_sb[0] + (1-weights_short) * mean_sb[1]
standard_deviation_short = np.sqrt(weights_short ** 2 * std_sb[0] ** 2 + (1 - weights_short) ** 2 * std_sb[1] ** 2 + 2 * weights_short * (1-weights_short) * cov_sb)
plt.plot(standard_deviation, expected_return, linestyle='-', c='red')
plt.plot(standard_deviation_short, expected_return_short, linestyle='--', c='red')
plt.plot(std_tan, mean_tan, marker="o", markersize=3, markeredgecolor="blue")
plt.annotate(" tan", (std_tan, mean_tan))
plt.plot(std_6040, mean_6040, marker="o", markersize=3, markeredgecolor="green")
plt.annotate(" 60/40", (std_6040, mean_6040))
plt.plot(std_sb[0], mean_sb[0], marker="o", markersize=3, markeredgecolor="yellow")
plt.annotate(" s", (std_sb[0], mean_sb[0]))
plt.plot(std_sb[1], mean_sb[1], marker="o", markersize=3, markeredgecolor="orange")
plt.annotate(" b", (std_sb[1], mean_sb[1]))
plt.xlabel('standard deviation')
plt.ylabel('expected return')
plt.title('mean-standard deviation frontier')
plt.grid(True)
plt.show
```

Q:What explains the difference between the RP and RP- unlevered portfolio performance?

To explain the difference between the unlevered and levered RP portfolios performance, we must consider the construction of the RP portfolio. The RP portfolio is diversified by risk instead of value: the idea is to have similar amounts of risk in stocks and in bonds, which is why the weights are chosen proportional to the inverse of standard deviations. For this reason, more is invested in low risk assets (bonds) than in high risk assets (stocks). Thus, the risk of the portfolio has to be increased (using leverage) to levels comparable to those of other portfolios (in this specific case, the 60/40 portfolio), and because the return per unit of risk is higher for RP strategies, when the risk is as high as most investors require, the expected return is much higher than the classic portfolios such as the 60/40 one or the market value weighted portfolio.  

## (c)
Q:Why does the RP strategy in Figure 2 of AFP not actually lie on the efficient frontier?

The RP strategy does not lie on the efficient frontier because the efficient frontier is by construction the minimum variance portfolios, which equalize the marginal contributions of each asset to total risk (focuses on allocation of capital), the RP portfolio equalizes each assets total risk contribution. In other words, minimum variance portfolios minimize the variance of a portfolio (no matter how the risk contributions of assets is distributed), while the RP portfolio constrains each asset's contribution so that all assets contribute equally to the overall risk. In the AFP paper, the RP portfolio is constructed using a three year rolling volatility of monthly excess returns in a way to rebalance the portfolio monthly so as to target an equal risk allocation across assets. So when they average all the weights for each calculation, the result is different than the one we find when considering the complete sample at once. While they use this construction for the RP portfolio, the efficient frontier is constructed using the whole sample at once, which is why the RP portfolio is not on the efficient frontier in the AFP analysis.

Although lower volatility can be acheived in the period after the 90s (efficient frontier is shifted to the left), the Sharpe ratio of the tangent portflio was lower in this second subsample (90s-2023). However we must note that we used the same risk free rate in both cases while they were actually higher in the 80s and 90s than they are now, so it is probably better not to focus on the Sharpe ratios themselves, but rather the fact that for the period before 1990, the levered RP is much further away from the CML than the one after 1990. This makes the RP portfolio more efficient in this second subsample, and this can be explained by the fact that, as noted before, the correlation between stocks and bonds was positive before the 90s and then turned negative after the 90s, making the RP portfolio much more efficient, because it is better diversified (the inverse-volatility weighting scheme ignores pairwise correlations), and for that it is closer to the CML than it is in the period before the 90s.
